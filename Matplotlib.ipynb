# Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.linspace(0,10,20)
print(x)
y = x*x
z = x+y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Demo graph')
```

```python
plt.plot(x,x)
```

```python
plt.plot(x,x,y,x,x,y,x,z)
```

```python
# Subplot- for plotting more than one graphs
# parameters:(row, column, no_of_graphs)
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(x,x)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Demo graph')
plt.tight_layout()
```

# Types of plot

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.hist(x,y)
plt.subplot(2,2,3)
plt.bar(x,y)
plt.subplot(2,2,4)
plt.fill(x,y)
plt.tight_layout()
```

```python
plt.polar(x,y)
```

```python
# Legend
```

```python
plt.plot(x,y,y,x,label = 'first example')
plt.legend(loc = 0)
plt.title("legends")
```

```python
# object oriented method of plotting


fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('Object oriented plotting using Figure')
axes.plot(x,y,y,x)
```

```python
# Multiple plotting


fig = plt.figure()
x1 = fig.add_axes([0.1,0.1,1,1])
x1.set_xlabel('x-axis')
x1.set_ylabel('y-axis')
x1.set_title("Graph1")
x1.plot(x,y)

x2 = fig.add_axes([0.4,0.4,0.4,0.4])
x2.set_xlabel('x-axis')
x2.set_ylabel('y-axis')
x2.set_title("Graph2")
x2.plot(x,y)
```

```python
# subplotting

fig, axes = plt.subplots(1,2) # subplotting
axes[0].plot(x,y)
axes[1].plot(x,y)

axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].set_title('Graph1 of row1')

axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].set_title('Graph2 of row1')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2,2) # 2d subplotting  

axes[(0,0)].set_title('Graph0 of row0')
axes[(1,1)].set_xlabel('x-axis')
axes[(1,1)].set_title('Graph1 of row1')
axes[(1,1)].plot(x,y)
axes[(0,1)].set_xlabel('x-axis')
axes[(0,1)].set_title('Graph1 of row0')
axes[(0,1)].plot(y,x)
plt.tight_layout()
```

```python
# Resizing and styling
```

```python
fig = plt.figure(figsize = (3,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'red',linewidth = 1,linestyle = '--', alpha = 0.5)
```

```python
fig, axes = plt.subplots(2,1, figsize=(3,5))
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
# styling: markers
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('Markers')
axes.plot(x,y, marker = 'o',markersize = '7',markerfacecolor='yellow',markeredgecolor='red',markeredgewidth= '2')
```

```python
# set limit
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.set_title('limits')
axes.plot(x,y)
axes.set_xlim([2,10])
axes.set_ylim([20,60])
```
